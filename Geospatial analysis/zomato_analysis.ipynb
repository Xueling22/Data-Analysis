```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
con=sqlite3.connect('Datasets/zomato_rawdata.sqlite')
```

```python
df=pd.read_sql_query("SELECT * FROM Users",con)
```

```python
df.isnull().sum()/len(df)*100
```

```python
df.replace({'rate': {'NEW': np.nan, '-': np.nan}}, inplace=True) #np.nan常被用来表示数据集中某些位置的值是未知的或不存在的
```

```python
df['rate'].unique() #只要前面的分不要/5
```

```python
df['rate'].apply(lambda x:float(x.split('/')[0]) if type(x)==str else x)
```

```python
x=pd.crosstab(df['rate'],df['online_order']) #交叉表（contingency table），它是用来显示（或汇总）两个（或更多）变量之间的频率分布的表格。
```

```python
#不用索引的形式因为这个crosstab有更多的功能和参数可以加，加索引实质上改变了dataFrame
```

```python
x.plot(kind='bar',stacked=True) #非100%堆积直方图
```

```python
normalize_df=x.div(x.sum(axis=1).astype(float),axis=0) #x.sum(axis=1)对水平行求和，.astype(float)转换浮点防止损失精度，div除
```

```python
(normalize_df*100).plot(kind='bar',stacked=True)
```

```python
df['rest_type'].isnull().sum()
```

```python
data=df.dropna(subset=['rest_type'])
```

```python
data['rest_type'].isnull().sum()
```

```python
quick_bites_df=data[data['rest_type'].str.contains('Quick Bites')]
```

```python
quick_bites_df.columns
```

```python
quick_bites_df['reviews_list']=quick_bites_df['reviews_list'].apply(lambda x:x.lower()) #忽略大小写
```

```python
from nltk.corpus import RegexpTokenizer
```

```python
tokenizer=RegexpTokenizer("[a-zA-Z]+")
```

```python
tokenizer
```

```python
tokenizer.tokenize(quick_bites_df['reviews_list'][3])
```

```python
sample=data[0:10000]
```

```python
reviews_tokens=sample['reviews_list'].apply(tokenizer.tokenize)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop=stopwords.words('english')
```

```python
stop.extend(['rated',"n","nan","x","RATED","Rated"]) #把这些也加进去，当作stop词，单双引号一样作用
```

```python
rev3=reviews_tokens[3]
```

```python
reviews_tokens_clean=reviews_tokens.apply(lambda each_review:[token for token in rev3 if token not in stop])
```

```python
reviews_tokens_clean
```

```python
total_reviews_2D=list(reviews_tokens_clean)
```

```python
total_reviews_1D=[] #转成Unigram
for review in total_reviews_2D:
    for word in review:
        total_reviews_1D.append(word)
        
```

```python
total_reviews_1D
```

```python
from nltk import FreqDist
```

```python
fd=FreqDist() #FreqDist是用来计数词项出现的次数，是一个类似字典的对象
```

```python
for word in total_reviews_1D:
    fd[word]= fd[word]+1  #如果词已经在fd中，就将其计数增加1；如果还没有出现过，就在fd中为其创建一个条目，并将计数设为1
```

```python
fd.most_common(20)
```

```python
fd.plot(20)
```
